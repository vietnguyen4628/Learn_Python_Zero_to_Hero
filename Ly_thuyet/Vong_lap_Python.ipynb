##  1. Giới thiệu về vòng lặp
Vòng lặp giúp chạy lặp lại đoạn mã nhiều lần. Python có 2 loại vòng lặp:
- `for`: biết trước số lần lặp
- `while`: lặp đến khi điều kiện sai


## 2. Vòng lặp `for`
### Cú pháp:
```python
for biến in range(bắt_đầu, kết_thúc, bước):
    # code
```


````python
# Ví dụ 1: In từ 1 đến 10
for i in range(1, 11):
    print(i, end = " ")
````

````python
# Ví dụ 2: Tính tổng 1..n
n = int(input("Nhập vào n: "))
tong = 0
for i in range(1, n+1):
    tong += i
print('Tổng =', tong)
````

## 3. Vòng lặp `while`
### Cấu trúc:
```python
while điều_kiện:
    # code
```


````python
# Ví dụ: nhập đến khi nhập 0
x = 3
while x != 0:
    print('Bạn nhập:', x)
    x -= 1
````

## 4. Kỹ thuật nâng cao
### 4.1. break
Chức năng: Dừng hẳn vòng lặp ngay lập tức, kể cả điều kiện lặp vẫn còn đúng.

Khi dùng:
- Tìm thấy kết quả mong muốn và muốn thoát sớm.
- Ngăn vòng lặp chạy không cần thiết.


````python
#Ví dụ: Tìm số đầu tiên chia hết cho 7 trong 100 số đầu tiên

for i in range(1, 101):
    if i % 7 == 0:
        print("Số đầu tiên chia hết cho 7:", i)
        break
````

### 4.2. continue

Chức năng: Bỏ qua phần còn lại của vòng lặp trong lần lặp hiện tại và chuyển sang lần tiếp theo.

Khi dùng:
- Muốn bỏ qua một số trường hợp không cần xử lý.
- Giúp code gọn hơn thay vì viết logic lồng nhau.

````python
#Ví dụ: In các số từ 1 đến 10, nhưng bỏ qua số chẵn

for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
````

### 4.3. nested loop (vòng lặp lồng nhau)

Chức năng: Một vòng lặp nằm bên trong một vòng lặp khác.

Thường dùng trong:
- Xử lý ma trận
- Thuật toán sắp xếp
- Vẽ pattern, xử lý tổ hợp


````python
#Ví dụ: In ma trận 3x3

for i in range(3):
    for j in range(3):
        print(f"({i}, {j})", end=" ")
    print()
````

## 5. Bài tập
#### Bài 1: In các số từ 1 -> n. Với n nhập vào là số nguyên
#### Bài 2: Tính tổng các số chẵn từ 1 -> n. Với n nhập vào là số nguyên
#### Bài 3. Đảo số

````python
# Bài 1: In các số từ 1 -> n. Với n nhập vào là số nguyên

def main():
    n = int(input("Nhập vào n: "))

    for i in range(1, n + 1):
        print(i, end = " ")
main()
````

````python
# Bài 2: Tính tổng các số chẵn từ 1 -> n. Với n nhập vào là số nguyên
````

````python
# Bài 3: Đảo số. Với n là số nguyên được nhập vào từ bàn phím

def main():
    n = int(input("Nhập số nguyên n: "))

    sign = -1 if n < 0 else 1
    n = abs(n)

    rev = 0
    while n > 0:
        digit = n % 10          # lấy chữ số cuối
        rev = rev * 10 + digit  # ghép vào số đảo
        n //= 10                # bỏ chữ số cuối

    rev *= sign                 # gắn lại dấu (nếu âm)

    print("Số đảo là:", rev)

main()
````

````python
# Bài 3: Đảo số. Với n là số nguyên được nhập vào từ bàn phím

def main():
    n = int(input("Nhập số nguyên n: "))

    sign = -1 if n < 0 else 1
    n = abs(n)

    rev = ''

    # xây chuỗi đảo
    if n == 0:
        rev = '0'
    else:
        while n > 0:
            digit = n % 10          # lấy chữ số cuối
            rev += str(digit)  # ghép vào số đảo
            n //= 10                # bỏ chữ số cuối

        # gắn lại dấu nếu âm
        if sign == -1:
            rev = "-" + rev

        print("Số đảo là:", rev)

main()
````
