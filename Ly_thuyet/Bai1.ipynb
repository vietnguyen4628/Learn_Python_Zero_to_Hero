**Bài 1: Cách tạo ghi chú trong python**

**Mục đích của việc ghi chú là gì?**
- Giải thích được mục tiêu của đoạn code
- Ghi lại ý tưởng của đoạn code
- Làm cho một dòng / đoạn code sẽ không hoạt động
**Cách ghi chú trong python:**
- Ghi chú 1 dòng: Sử dụng dấu # để ghi chú 1 dòng
- Ghi chú nhiều dòng/ đoạn mã: sử dụng ''' nội dung cần được chú thích '''
